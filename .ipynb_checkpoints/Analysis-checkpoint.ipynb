# 1. <a id='section_1'>Measures</a>
<a href='#TOC'>Back to table of Contents</a>

In the file clean_data.csv each row corresponds to an individual golfer and their statistics collected for a particular PGA tour season. Find the columns below:

| #   | Measure                    | Explanation                                                                                   |
| :--- | :-------------------------- | :--------------------------------------------------------------------------------------------- |
| 1   | ROUNDS                     | Number of PGA tour rounds played by a player in a specific year.                               |
| 2   | SCORING                    | Average score per round played for the year, adjusted for field strength.                       |
| 3   | DRIVE_DISTANCE             | Average yards per measured drive, reflecting a player's driving distance.                           |
| 4   | FWY_% (Fairway Percentage) | Percentage of tee shots that land in the fairway, indicating accuracy off the tee.            |
| 5   | GIR_% (Greens in Reg.)     | Percentage of greens hit in regulation, measuring accuracy of approach shots.                      |
| 6   | SG_P (Strokes Gained Putting)| Measures putting accuracy and proficiency by comparing putts taken to a baseline.                  |
| 7   | SG_TTG (Strokes Gained Tee to Green)| Per-round average of strokes gained or lost compared to field average, excluding putting.     |
| 8   | SG_T (Strokes Gained Total) | Per-round average of strokes gained or lost compared to field average, including all aspects.  |
| 9   | POINTS                     | Cumulative points earned in the FedEx Cup points race for the year, reflecting overall performance. |
| 10  | TOP 10                     | Number of top-10 finishes during the year, showing frequency of high placements.                 |
| 11  | 1s (Wins)                  | Number of tournament wins during the year, indicating first-place finishes.                       |
| 12  | Year                       | The year of the PGA tour season for the data entry.                                                   |
| 13  | MONEY                      | PGA tour earnings from the specified season, reflecting financial success.                        |
| 14  | COUNTRY                    | The player's home country or nationality.                                                        |

The following code calculates and compares the number of wins (1s) for different percentiles (1% -15%) of DRIVE_DISTANCE and SG_P. It then creates a DataFrame called results_df to store these results and indicates whether DRIVE_DISTANCE or SG_P had more wins for each percentile. The results are displayed in tabular form.

```python
# Sort dfnew by 'DRIVE_DISTANCE' and 'SG_P' in descending order
dfnew_sorted_dd = dfnew.sort_values(by='DRIVE_DISTANCE', ascending=False)
dfnew_sorted_sgp = dfnew.sort_values(by='SG_P', ascending=False)

# Initialize lists to store results
results = []

# Loop through desired percentiles (10% to 26%)
for top in range(10, 26, 5):
    top_n_percent = int(len(dfnew_sorted_dd) * (top / 100))

    # Extract the top n percent rows based on DRIVE_DISTANCE and SG_P
    top_n_percent_dd = dfnew_sorted_dd.iloc[:top_n_percent]
    top_n_percent_sgp = dfnew_sorted_sgp.iloc[:top_n_percent]

    # Calculate the number of wins for each percentile
    wins_dd = top_n_percent_dd['1s'].sum()
    wins_sgp = top_n_percent_sgp['1s'].sum()

    # Determine which parameter has more wins
    more_wins = 'Drive Distance' if wins_dd > wins_sgp else 'SG_P'

    # Append the results to the list
    results.append([top, wins_dd, wins_sgp, more_wins])

# Create a DataFrame from the results list
results_df = pd.DataFrame(results, columns=['Top %', 'Drive Distance Wins', 'SG_P Wins', 'More Wins'])

results_df
```

For top 10% SG_P values, the total number of wins a player gets = 76.

## t-tests

### Effect on POINTS

```python
from scipy import stats

# Split the data into two groups based on a threshold (e.g., median)
threshold = df['DRIVE_DISTANCE'].median()
group1 = df[df['DRIVE_DISTANCE'] < threshold]['POINTS']
group2 = df[df['DRIVE_DISTANCE'] >= threshold]['POINTS']

# Perform a two-sample t-test
t_statistic, p_value = stats.ttest_ind(group1, group2)

if p_value < 0.05:
    print("There is a significant difference in POINTS based on DRIVE_DISTANCE.")
else:
    print("There is no significant difference in POINTS based on DRIVE_DISTANCE.")
```

```python
# Split the data into two groups based on a threshold (you can use any threshold)
threshold = df['SG_P'].median()
group1 = df[df['SG_P'] < threshold]['POINTS']
group2 = df[df['SG_P'] >= threshold]['POINTS']

# Perform a two-sample t-test
t_statistic, p_value = stats.ttest_ind(group1, group2)

if p_value < 0.05:
    print("There is a significant difference in POINTS based on SG_P.")
else:
    print("There is no significant difference in POINTS based on SG_P.")
```

### ANOVA tests

### Effect on POINTS

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Create an ANOVA model for DRIVE_DISTANCE
model_drive_distance = ols('POINTS ~ DRIVE_DISTANCE', data=df).fit()

# Perform the ANOVA
anova_table_drive_distance = sm.stats.anova_lm(model_drive_distance, typ=2)

# Check if there are significant differences among groups
if anova_table_drive_distance['PR(>F)'][0] < 0.05:
    print("There is a significant difference in POINTS based on DRIVE_DISTANCE.")
else:
    print("There is no significant difference in POINTS based on DRIVE_DISTANCE.")
```

```python
# Create an ANOVA model
model = ols('POINTS ~ SG_P', data=df).fit()

# Perform the ANOVA
anova_table = sm.stats.anova_lm(model, typ=2)

# Check if there are significant differences among groups
if anova_table['PR(>F)'][0] < 0.05:
    print("There is a significant difference in POINTS based on SG_P.")
else:
    print("There is no significant difference in POINTS based on SG_P.")
```

### Effect on SCORING

```python
# Create an ANOVA model
model = ols('SCORING ~ DRIVE_DISTANCE', data=df).fit()

# Perform the ANOVA
anova_table = sm.stats.anova_lm(model, typ=2)

# Check if there are significant differences among groups
if anova_table['PR(>F)'][0] < 0.05:
    print("There is a significant difference in SCORING based on DRIVE_DISTANCE.")
else:
    print("There is no significant difference in SCORING based on DRIVE_DISTANCE.")
```

```python
# Create an ANOVA model
model = ols('SCORING ~ SG_P', data=df).fit()

# Perform the ANOVA
anova_table = sm.stats.anova_lm(model, typ=2)

# Check if there are significant differences among groups
if anova_table['PR(>F)'][0] < 0.05:
    print("There is a significant difference in SCORING based on SG_P.")
else:
    print("There is no significant difference in SCORING based on SG_P.")
```

## Segmentation approach

### Lasso Regression

Lasso Regression is used for various reasons, and its benefits include:

* Picking the Best Features: Lasso helps us find which factors matter most by ignoring some that don't matter much.

* Easy to Understand: It gives us a model that's not too complicated, so we can easily see how each thing affects the result.

* Avoiding Confusion: When some things are very similar, Lasso picks just one, avoiding confusion.

* Not Overdoing It: Lasso prevents our model from trying too hard to match the data exactly, which is often not a good idea.

* Sorting by Importance: It tells us which factors are most important by showing their size.

* Lasso makes it easier to figure out what's important, keeps things simple, and prevents our model from being too obsessed with the data.

We are using unscaled data for this algorithm.

```python
# Instantiate the Lasso Regression model
lasso = Lasso(alpha=0.01)  # You can adjust the alpha (regularization strength) as needed

# Fit the model to the data
lasso.fit(X, y)

# Predict the target variable 'y' using the trained model
y_pred = lasso.predict(X)

# Calculate the R-squared score
r2 = r2_score(y, y_pred)

# Print the R-squared score
print("R-squared (R^2) Score:", r2)

# Get the coefficients (importance) of the features
coefficients = lasso.coef_

# Create a DataFrame to store feature names and their coefficients
feature_importance = pd.DataFrame({'Feature': X.columns, 'Coefficient': coefficients})

# Sort the DataFrame by coefficient values (importance)
feature_importance = feature_importance.sort_values(by='Coefficient', ascending=False)

# Create a bar chart to visualize the feature importance
plt.figure(figsize=(8, 6))
plt.barh(feature_importance['Feature'], feature_importance['Coefficient'], color='lightgreen')
plt.xlabel('Coefficient Value')
plt.title('Lasso Regression Feature Importance')
plt.gca().invert_yaxis()  # Invert the y-axis to display most important features at the top
plt.show()
```

This shows that SG_P is more important than DRIVE_DISTANCE for this algorithm with SG_TTG being the most important.

## Random Forest: 
* Random Forest is known for its high prediction accuracy. 
* It can handle both scaled and unscaled data.
* It combines the predictions of multiple decision trees. By aggregating the predictions of these individual trees, Random Forest can produce more accurate and robust predictions compared to a single decision tree or other models. 
* Other reasons why RF was chosen is because it is also robust to overfitting and has reduced risk of bias. Why? Because RF is an ensemble method. Each decision tree is trained on different subsets of the data and features, and predictions of each tree are combined to get accurate results.Hence, risk of systematic errors or biases is avoided.
* It can pick up on any complex relationships between variables, including nonlinear and interaction effects.
* It also provides a measure of feature importance and indicates which variables are most influential in making predictions making it good for feature selection.

```python
X = dfnew[['DRIVE_DISTANCE','SG_P', 'FWY_%', 'GIR_%', 'SG_TTG']]
y = dfnew['1s'].values.ravel()  # Flatten 'y' to a 1D array

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE

# Set seed for reproducibility
SEED = 1

# Split dataset into 70% train and 30% test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=SEED)

# Instantiate a random forests regressor 'rf' with 400 estimators
rf = RandomForestRegressor(n_estimators=400, min_samples_leaf=0.12, random_state=SEED)

# Fit 'rf' to the training set
rf.fit(X_train, y_train)

# Predict the test set labels 'y_pred'
y_pred = rf.predict(X_test)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)

# Calculate Mean Squared Error (MSE)
rmse = mean_squared_error(y_test, y_pred)**(0.5)

# Calculate R-squared (R^2)
r2 = r2_score(y_test, y_pred)

# Print the metrics
print('Mean Absolute Error (MAE): {:.2f}'.format(mae))
print('Root Mean Squared Error (RMSE): {:.2f}'.format(rmse))
```

MAE:
* MAE is the average of the absolute differences between predicted and actual values. 
* It tells you, on average, how far off your predictions are from the true values. 
* Smaller MAE values indicate better model performance, as they suggest that the model's predictions are closer to the actual values.

RMSE:
* Lower RMSE values indicate better model performance because they mean that the model's predictions are, on average, closer to the actual values.
* RMSE is expressed in the same units as the target variable, making it interpretable.

```python
threshold = 0.5  # can be adjusted

# Convert predicted probabilities to binary predictions
y_pred_binary = (y_pred > threshold).astype(int)

# Calculate and print accuracy
accuracy = accuracy_score(y_test, y_pred_binary)
print('Accuracy:', accuracy)
```

Accuracy is a metric that measures the overall correctness of the model's predictions.

```python
# Create a pd.Series of feature importances
importances_rf = pd.Series(rf.feature_importances_, index=X.columns)

# Sort importances_rf
sorted_importances_rf = importances_rf.sort_values()

# Make a horizontal bar plot
plt.figure(figsize=(8, 6))
sorted_importances_rf.plot(kind='barh', color='lightgreen')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Random Forest Feature Importance')
plt.show()
```

This also shows that SG_P is more important than DRIVE_DISTANCE for this algorithm with SG_TTG being the most important.

### Tuning RF Hyperparameters

Here we find the best parameters for our RF model.

```python
SEED = 1

# Instantiate a random forests regressor 'rf' 
rf = RandomForestRegressor(random_state= SEED)

# Inspect rf' s hyperparameters
# rf.get_params()

from sklearn.metrics import mean_squared_error as MSE
from sklearn.model_selection import GridSearchCV
# Define a grid of hyperparameter 'params_rf'
params_rf = {'n_estimators': [300, 400, 500],
             'max_depth': [4, 6, 8],
             'min_samples_leaf': [0.1, 0.2],
             'max_features': ['log2', 'sqrt']
            }

# Instantiate 'grid_rf'
grid_rf = GridSearchCV(estimator=rf,
                       param_grid=params_rf,
                       cv=3,
                       scoring='neg_mean_squared_error',
                       verbose=1,
                       n_jobs=-1)

# Fit 'grid_rf' to the training set (search best hyperparams)
grid_rf.fit(X_train, y_train)

# Extract the best hyperparameters from 'grid_rf'
best_hyperparams = grid_rf.best_params_
print('Best hyperparameters:\n', best_hyperparams)
```

### Evaluating the best model performance

```python
# Extract the best model from 'grid_rf'
best_model = grid_rf.best_estimator_
# Predict the test set labels
y_pred = best_model.predict(X_test)
# Evaluate the test set RMSE
rmse_test = MSE(y_test, y_pred)**(1/2)
# Print the test set RMSE
print('Test set RMSE of RF model: {:.2f}'.format(rmse_test))
```

```python
# Extract feature importances from the best model
importances_rf = best_model.feature_importances_

# Create a DataFrame to store feature importances along with their names
feature_importance_df = pd.DataFrame({'Feature': X_train.columns, 'Importance': importances_rf})

# Sort the DataFrame by importance in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Create a horizontal bar plot for feature importances
plt.figure(figsize=(8, 6))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'], color='lightgreen')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Random Forest Feature Importance')
plt.gca().invert_yaxis()  # Invert the y-axis to have the most important features at the top
plt.show()
```

This also shows that SG_P is more important than DRIVE_DISTANCE for this algorithm with SG_TTG being the most important.
